```python
import pandas as pd
```

```python
## ЗАДАНИЕ 1 ##

data = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

df = pd.DataFrame(data.merge(movies))

def class_rating(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    if 2 < rating <= 4:
        return 'средний рейтинг'
    if 4.5 <= rating <= 5:
        return 'высокий рейтинг'

df['class'] = df['rating'].apply(class_rating)
df.head(250) 
```

```python
## ЗАДАНИЕ 2 ##

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

keywords = pd.read_csv('keywords.csv')

def region(value):
    for region, location in geo_data.items():
        for city in location:
            if city in value.lower():                
                return region
    return 'undefined'


keywords['region'] = keywords.keyword.apply(region)

keywords[ keywords['region'] == 'Северо-Запад' ].head()
```

```python
## ЗАДАНИЕ 3 ##

def check_year(movie):
    for year in list(range(1950,2011)):
        if str(year) in movie:
            return year
    return 1900

df['year'] = df['title'].apply(check_year)

df.head(10)
```
